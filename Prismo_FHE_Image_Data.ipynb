```python
from Pyfhel import Pyfhel, PyPtxt, PyCtxt
import pickle
import numpy as np
from PIL import Image
import io
import requests
import zlib
from IPython.display import display
```

# Initialize Pyfhel

```python
HE = Pyfhel()
HE.contextGen(scheme='BFV', n=2**14, t_bits=20)
HE.keyGen()
```

# Public Key, Secret Key

```python
print(HE)
```

# Download image

```python
image_url = "https://raw.githubusercontent.com/ALGOREX-PH/LeafGuard-AI/refs/heads/main/src/images/LeafGuard_AI_Logo.png"
output_image_path = "LeafGuard_AI_Logo.png"
response = requests.get(image_url, stream=True)
with open(output_image_path, 'wb') as file:
    for chunk in response.iter_content(chunk_size=8192):  # Iterate over data in chunks
        file.write(chunk)  # Write each chunk to the file
```

# Load Image

```python
img = Image.open("/content/donut_pixel_art_16x16_by_sirensence_dgi67hn-fullview.jpg")
```

# Turn Image to Numpy Array

```python
img_array = np.array(img)
```

# Serialize the Numpy Array into Text

```python
serialized = pickle.dumps(img_array)
#compressed_data = zlib.compress(pickle.dumps(img_array))
```

# Text

```python
Text = serialized.hex()
```

```python
print(Text)
```

# Encryption

```python
encoded_text = [ord(char) for char in Text]
```

```python
cipher_text = [HE.encryptInt(np.array([num], dtype=np.int64)) for num in encoded_text]
print(f"Cipher Text: {[str(ct) for ct in cipher_text]}")
```

```python
decrypted_text = [HE.decryptInt(ct)[0] for ct in cipher_text]
decoded_text = ''.join([chr(num) for num in decrypted_text])

print(f"Original Text: {Text}")
print("")
print(f"Decoded Text after encryption/decryption: {decoded_text}")
```

```python
decoded_bytes = bytes.fromhex(decoded_text)
```

```python
img_array = pickle.loads(decoded_bytes)
```

```python
img = Image.fromarray(img_array)
```

```python
img.save("Decoded_Image.jpg")
```
