# Import Packages

```python
from Pyfhel import Pyfhel, PyPtxt, PyCtxt
import pickle
import numpy as np
import pandas as pd
```

```python
HE = Pyfhel()
HE.contextGen(scheme='BFV', n=2**14, t_bits=20)
HE.keyGen()
```

```python
print(HE)
```

```python
file_path = '/content/Tables.xlsx'
df = pd.read_excel(file_path)
serialized_data = pickle.dumps(df)
```

```python
print(serialized_data)
```

```python
XLSX = serialized_data.hex()
```

```python
print(XLSX)
```

```python
encoded_text = [ord(char) for char in XLSX]
```

```python
cipher_text = [HE.encryptInt(np.array([num], dtype=np.int64)) for num in encoded_text]
print(f"Cipher Text: {[str(ct) for ct in cipher_text]}")

decrypted_text = [HE.decryptInt(ct)[0] for ct in cipher_text]
decoded_text = ''.join([chr(num) for num in decrypted_text])
print(f"Original Text: {XLSX}")
print(f"Decoded Text after encryption/decryption: {decoded_text}")
```

```python
decoded_bytes = bytes.fromhex(decoded_text)
```

```python
deserialized_df = pickle.loads(decoded_bytes)
```

```python
deserialized_df.to_excel('Tables_Decrypted.xlsx', index=False)
```
