```python
from Pyfhel import Pyfhel, PyPtxt, PyCtxt
import pickle
import numpy as np
```

# Initialize Pyfhel

```python
HE = Pyfhel()
HE.contextGen(scheme='BFV', n=2**14, t_bits=20)
HE.keyGen()
```

# Public Key, Secret Key

```python
print(HE)
```

# Serialize PDF File

```python
try:
    # Read the PDF file in binary mode
    with open("/content/financial-wellness-journal-english.pdf", 'rb') as pdf_file:
        pdf_data = pdf_file.read()

    # Serialize the binary data using pickle
    with open("PDF.pkl", 'wb') as pickle_file:
        serialized_pdf = pickle.dumps(pdf_data)

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
PDF = serialized_pdf.hex()
```

```python
print(PDF)
```
